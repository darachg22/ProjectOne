```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Here I had imported the necessary libraries to create the code. I used numpy for the matematical operation and pandas for data analytical options. I have used both number of times during my course. Following this, when analysing the code, I ass Matplotlib and Seaborn to display the findings in order to study them.

```python
number_of_students = 28
```

Input for the number of students for the data. I have decided to begin with the same number in the real life survey for some semblance of symmetry. 

```python
student_numbers = np.arange(1, number_of_students + 1)
```

Here I considered the best way to create the list of student data. I considered using the faker library at first but then decided that my results could be achieved by creating numbers and the allocating different points to these numbers. The best way to this was to use numpy. I had spent a bit of time trying to create this section of the code with the '+ 1' at the end of the code solving a number of issues with the output. I found how to use the arrange function effectively here: https://numpy.org/doc/stable/reference/generated/numpy.arange.html.

```python
average_grade_mean = 58
average_grade_dev = 15
average_grade = np.random.normal(average_grade_mean, average_grade_dev, number_of_students).astype(int)
```

Here I needed to create code that fabricates different average grades for the simulation. Once again I had consdiered different ways to but the numpy random function was of great use here, and throughout this project. Once ran I noticed that the grades were too random at first so I began looking at creating an overall average and standard deviation to better simulate real life grades. My thinking was the real life results varied altough not by the amounts the simulated data did. I added these sections to counteract this. Once agin I recived this information from the numpy manual: https://numpy.org/doc/stable/reference/random/generated/numpy.random.normal.html, https://numpy.org/doc/stable/reference/random/index.html#quick-start.

```python
probability_College = 0.81
going_to_third_level = np.random.choice([1, 0], number_of_students, p=[probability_College, 1 - probability_College])
```

Considering the task at hand I realised that I could use similar code for the next two sections i.e the third level education and the part time job sections. I also considered the use of probability and how I could use the real life study to find a probaility to enter into the code. I used the numpy choice function but I also considered using a boolean fucntion but found it did not fit my parameters. As the choice function uses a 1-D array I found this the best way to achieve the results I wanted. With the choice function I then added the options of using 1 to represent the student intending to go to third level and 0 representing the student not going to third level. I learned an inroduction to this here: https://www.geeksforgeeks.org/numpy-random-choice-in-python/. I struggled with syntax of this section of code and needed to look into how to use this to get the result I wanted. I came across this site which showed me how to use the 'p=' elements to get the required results: https://sparkbyexamples.com/python/how-to-use-numpy-random-choice-in-python/#google_vignette. I then inputted the probability from the real life study.

```python
probability_job = 0.5
part_time_job = np.random.choice([1, 0], number_of_students, p=[probability_job, 1 - probability_job])
```

To achieve the next section of code I just needed to recreate the code from the last section as they will both serve to provide different information but use the same structure.  

```python
study_mean = 1.5
study_std = 1
study_hours = np.random.normal(study_mean, study_std, number_of_students).astype(int)
```

Similarily to previous section of code, here I needed to create study hours for the students. Once again I wanted to use previous code to help with . The only major difference being when I first ran the code it showed a number of minus values. To combat this I came across the numpy absolute function which tidied this up. The information I needed to execute this effectively was here: https://sparkbyexamples.com/python/numpy-absolute-value/?expand_article=1. I then needed to change the float number to integers. This was a problem I faced in a previous assignment and I was led to '.astype' to fix the problem: https://www.geeksforgeeks.org/python-pandas-dataframe-astype/. I then added this to all the sections that did not need float numbers. 

```python
School_view_mean = 5
School_view_std = 2
overall_view_school = np.abs(np.random.normal(School_view_mean, School_view_std, number_of_students)).astype(int)
```

Once again I was lucky that a repitition of sections was necessary to achieve the outcomes. Here I have created a section of code that creates the students feelings towards school and their final year. Once again I have inputted the mean and deviatation from the real life survey.

```python
gender = np.random.choice([0, 1], number_of_students)
```

Here I created a code that creates assigns a gender to the students in order to derive further information. Here 0 represents male and 1 represents female. 

```python
ages = np.random.randint(16, 20, number_of_students)
```

Finally I added an age section to the code. I set this so it creates a random age between 16 and 18 coinciding with the real life survey.

```python

data = {
    'Student ID': student_numbers,
    'Average Grade': average_grade,
    'Going to Third Level Education': going_to_third_level,
    'Part-time Job': part_time_job,
    'Study Hours per night': study_hours,
    'Overall View of School': overall_view_school,
    'Age': ages,
    'Gender': gender
}
```

Here I placed all the information into a data frame in order to see the results. Below the function to display the data through pandas is shown: 

```python
df = pd.DataFrame(data)

df 
```

# Displaying the Data Frame

Below is the code for the plotting of the dataframe. I have used a number of different plots to visualise the data. I had knowledge of how to do this from previous assignments and used this to create the plots. Once I had experimented with a number of different plots I decided on these as conatining the most useful information to the four questions mentioned in the introduction. When testing and exploring various plots I faced a number of challenges in particular making sure each was displayed clearly and functionally. I have decided to break the plots into two sections. The first being general data from the dataframe and the second showing the correlations between the results and plots that seek to show the information from the four questions mentioned in the introduction of this project.

Resources used:
Seaborn: https://seaborn.pydata.org/tutorial/introduction.html

Matplotlib: https://www.w3schools.com/python/matplotlib_pyplot.asp 

```python
plt.figure(figsize=(18, 6))

#1.Grades
plt.subplot(1, 4, 1)
sns.histplot(df['Average Grade'], bins=15, kde=True, color='skyblue')
plt.title('Distribution of Average Grade')
plt.xlabel('Average Grade')
plt.ylabel('Frequency')

#2.Third-Level
plt.subplot(1, 4, 2)
sns.countplot(x='Going to Third Level Education', data=df)
plt.title('Count of Students Going to Third Level Education')

#3.Part-time Job
plt.subplot(1, 4, 3)
sns.countplot(x='Part-time Job', data=df)
plt.title('Count of Students with Part-time Job')

#4.Gender
plt.subplot(1, 4, 4)
sns.countplot(x='Gender', data=df)
plt.title('Count of Students by Gender')

plt.tight_layout()

#Correlative visualisations
plt.figure(figsize=(18, 6))

#5.Study v Grades v Third-level
plt.subplot(2, 3, 2)
sns.scatterplot(x='Study Hours per night', y='Average Grade', data=df, hue='Going to Third Level Education')
plt.title('Study Hours per Night vs Average Grade')

#6.Age distribution
plt.subplot(2, 3, 3)
sns.histplot(df['Age'], bins=20, kde=True)
plt.title('Age Distribution')

#7.Study x Third-level
plt.subplot(2, 3, 4)
sns.boxplot(x='Going to Third Level Education', y='Study Hours per night', data=df)
plt.title('Study Hours per Night by Third Level Education')

#8.Third-level x View of School
plt.subplot(2, 3, 5)
sns.boxplot(x='Going to Third Level Education', y='Overall View of School', data=df)
plt.title('Overall View of School by Third Level Education')

#9.Age x Grade
plt.subplot(2, 3, 6)
sns.scatterplot(x='Average Grade', y='Age', data=df, hue='Going to Third Level Education')
plt.title('Average Grade vs Age')



plt.tight_layout()
plt.show()
```
